```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray
from skimage.transform import resize
import cv2
from scipy import ndimage
from skimage.util import random_noise
import os
import torch
from tqdm import tqdm
```

```python
class DigitsGeneration:
    
    def __init__(self, path_to_digits_templates, path_to_save):
        
        self.path_to_save = path_to_save
        self.images = []
        # self.tensor_status = False
        self.CLASSES = []
        self.IND = 0
        
        for i in range(10):
            img = plt.imread(path_to_digits_templates+f'{i}.png')
            img = rgb2gray(img)
            img = resize(img,(90,60))
            T = threshold_otsu(img)
            self.images.append(img<T)
            
    @staticmethod
    def crop(img, left_px=0, right_px=0):
        h, w = np.array(np.where(img))
        h_max, h_min = h[np.argmax(h)], h[np.argmin(h)]
        return img[h_min:h_max, left_px:img.shape[1]-right_px]
    
    def generate_strip_of_digits(self,  tresh=20):
        self.starts = []
        self.end_of_images = []
        self.start_of_images = []
        for i, img in enumerate(self.images + [self.images[0]]):
            
            if i==0:      
                long_image = np.zeros((tresh, img.shape[1]))
                self.starts.append(0)
                
            self.start_of_images.append(long_image.shape[0])
            
            long_image = np.concatenate((long_image, self.crop(img)), axis=0)    
            end = long_image.shape[0]
            long_image = np.concatenate((long_image, np.zeros((tresh, img.shape[1]))), axis=0)
            
            if i!=10:
                self.starts.append(long_image.shape[0] - tresh//2)
                
            self.end_of_images.append(end)
            
        self.long_image = long_image
        
        
    def divide_strip(self, stripe):
        
        self.images_divided = []
        self.classes = []
        
        for i in range(len(self.starts)-1):
            separators = np.linspace(self.starts[i], self.starts[i+1], 30, dtype=np.int32, endpoint=False)

            for sep in separators:
                self.images_divided.append(stripe[sep:sep+65,:])

                c1 = 0
                while True:
                    if (sep < self.end_of_images[c1]) or c1 == 10:
                        break
                    else:
                        c1+=1              
                c2 = 0
                while True:
                    if (sep+65 <= self.start_of_images[c2]) or c2 == 10:
                        break
                    else:
                        c2+=1
                c2-=1
                self.classes.append(f'{c1}_{c2}')

        cl = ['0_0', '0_1', '1_1', '1_2', '2_2','2_3','3_3','3_4','4_4','4_5','5_5','5_6','6_6','6_7','7_7','7_8','8_8','8_9','9_9', '10_9']
        cl_mapper = dict(zip(cl, range(len(cl))))
        self.classes = [cl_mapper[c] for c in self.classes]
        
    def image_transforms(self, img, cl): 
        '''Apply the following transforms to get more realistic examples
        #1: Morphology transforms - without/ with erosion of 2 degrees
        #2: Bluring - without/3 random sigma from interval operations
        #3: Moving from center without/left (2 ex) or right (2 ex)
        #4: Small rotations like 5, 10, -10, 5 degrees one right, one left/without
        #5: Noise gaussian , salt-pepper , poisson and speckle noise random (1 of each)/without
        # - 6: Geometrical transform: try to make the middle size of image closer (as in water meter)/without
        '''
    
        for i1 in range(3):
            img_ = img.copy().astype(np.float32)
            h,w = img_.shape
            if i1:
                img_ = cv2.morphologyEx(img, cv2.MORPH_ERODE, cv2.getStructuringElement(cv2.MORPH_RECT,(3,3)),iterations=i1)
            
            for i2 in range(5):
                if i2:
                    shift = np.sign(np.random.randint(-1,2))*np.random.randint(5,15)
                    if shift>0:
                        img_[:, shift:] = img_[:, :img_.shape[1]-shift]
                        img_[:, :shift] = 0
                    else:
                        img_[:, :img_.shape[1]+shift] = img_[:, -shift:]
                        img_[:, img_.shape[1]+shift:] = 0
                    
                    for i3 in range(3):
                        if i3:
                            angle = np.random.randint(-7, 7)
                            img_ = ndimage.rotate(img_, angle=angle, mode='nearest')
                            img_ = img_[:h, :w]
                        for i4 in range(3):
                            if i4:
                                sigma = np.random.normal(0,3)
                                img_ = blur = cv2.GaussianBlur(img, (5,5), sigma) 

                            for i5 in range(5):
                                amount = np.random.uniform(0,0.02)
                                if i5==1:
                                    img_ = random_noise(img_, mode='gaussian')
                                if i5==2:
                                    img_ = random_noise(img_, mode='pepper', amount=amount)
                                if i5==3:
                                    img_ = random_noise(img_, mode='s&p', amount=amount)
                                if i5==4:
                                    img_ = random_noise(img_, mode='speckle')
                                
                                np.save(f'{self.path_to_save}{self.IND}.npy', img_)
                                self.IND+=1
                                
                                self.CLASSES.append(cl)
#                                 img_ = torch.FloatTensor(img_)
                                # class_ = torch.tensor([cl])
     
                                # if self.tensor_status:
#                                     self.IMAGES = torch.cat((self.IMAGES, img_.unsqueeze(2)), dim=-1)
                                    # self.CLASSES = torch.cat((self.CLASSES, class_))
                                # else:
#                                     self.IMAGES = torch.FloatTensor(img_.unsqueeze(2))
                                    # self.CLASSES = class_
                                    # self.tensor_status = True

#                         for i in range(2):
#                             x = np.arange(-img.shape[0]//2, img.shape[0]//2).reshape((-1,1))
#                             plt.imshow(img*(-(x/(img.shape[0]//2))**2), cmap='Greys_r')

    def create_df(self):
        self.generate_strip_of_digits()
        self.divide_strip(self.long_image)
        for i in tqdm(range(len(self.classes))):
            self.image_transforms(gen.images_divided[i], cl=self.classes[i])
        # torch.save(self.IMAGES, 'DigitsSyn.pt')
        # torch.save(self.CLASSES, 'DigitsSynClasses.pt')
        np.save('DIGITS_SYN_TARGET.npy', np.array(self.CLASSES))
            
```
